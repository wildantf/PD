```python
df=pd.read_csv("Unas.csv")

for col in df.columns:
    print("===",col,"===")
    print("jumlah data keseluruhan ",df[col].count())
    print("rata-rata   ",df[col].mean())
    print("nila minimal ",df[col].min())
    print("Q1       ",df[col].quantile(0.25))
    print("Q2          ",df[col].quantile(0.5))
    print("Q3          ",df[col].quantile(0.75))
    print("Nilai Max   ",df[col].max())
    print("kemencengan","{0:.2f}".format(round(df[col].skew(),2)))
    print("kemencengan          " ,"{0:.6f}".format(round(df[col].skew(),6)))
    print("Standar Deviasi   ","{0:.2f}".format(round(df[col].std(),2)))
    print("Variansi         ","{0:.2f}".format(round(df[col].var(),2)))
```
